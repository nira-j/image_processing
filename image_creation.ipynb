```python
import cv2
import numpy as np

img=np.ones((600,800,3))

color1=(0,0,0)
color2=(0,255,0)
color3=(0,0,10)


triangle0=np.array([[200,250],[0,100],[0,300]])
triangle1=np.array([[150,50],[0,300],[300,300]])
triangle2=np.array([[300,100],[200,300],[500,300]])
triangle3=np.array([[480,70],[350,300],[600,300]])
triangle4=np.array([[680,100],[550,300],[800,300]])
triangle5=np.array([[600,100],[450,300],[700,300]])
triangle6=np.array([[700,200],[800,100],[800,300]])
rect1=np.array([[0,300],[380,300],[0,600]])
rect2=np.array([[450,300],[800,300],[800,600]])

cv2.fillConvexPoly(img,triangle0,color1)
cv2.fillConvexPoly(img,triangle1,color1)
cv2.fillConvexPoly(img,triangle2,color1)
cv2.fillConvexPoly(img,triangle3,color1)
cv2.fillConvexPoly(img,triangle4,color1)
cv2.fillConvexPoly(img,triangle5,color1)
cv2.fillConvexPoly(img,triangle6,color1)
cv2.fillConvexPoly(img,rect1,color2)
cv2.fillConvexPoly(img,rect2,color2)

cv2.circle(img,(380,80),40,color3,-1)


cv2.imshow("myimage",img)
cv2.waitKey()
cv2.destroyAllWindows()
```
