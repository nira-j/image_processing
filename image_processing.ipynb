```python
import cv2
import numpy as np

image1 = cv2.imread("pic1.jpg")
image2 = cv2.imread("pic2.jpg")

# image1
cv2.imshow("myimage",image1)
cv2.waitKey()
cv2.destroyAllWindows()

# image2
cv2.imshow("myimage",image2)
cv2.waitKey()
cv2.destroyAllWindows()

# crop image1
croped_img1=image1[: , 320:640]
print(croped_img1.shape)
cv2.imshow("myimage",croped_img1)
cv2.waitKey()
cv2.destroyAllWindows()

# # crop image2
croped_img2=image2[: , 1280:1600]
print(croped_img2.shape)
cv2.imshow("myimage",croped_img2)
cv2.waitKey()
cv2.destroyAllWindows()

# swapping 
image1[:,320:640],image2[:,0:320] = croped_img2,croped_img1

cv2.imshow("myimage",image1)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imshow("myimage",image2)
cv2.waitKey()
cv2.destroyAllWindows()

#combining two images
img = np.ones((640,1080,3))
img=np.hstack((croped_img1,croped_img2))
cv2.imshow("myimage",img)
cv2.waitKey()
cv2.destroyAllWindows()
```
